# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
df = sns.load_dataset('tips')
```

### Посмотрите на первые 5 строчек

```python
df.head()
```

### Узнайте сколько всего строчек и колонок в данных

```python
rows, columns = df.shape
print(f'Количество строк: {rows}')
print(f'Количество колонок: {columns}')
```

### Проверьте есть ли пропуски в данных

```python
df.isna()
```

### Посмотрите на распределение числовых признаков

```python
for column in df.columns:
    sns.histplot(data=df, x=column)
    plt.title(f'Гистограмма для признака "{column}"')
    plt.show()
```

### Найдите максимальное значение 'total_bill'

```python
total_bill_max = df['total_bill'].max()
total_bill_max
```

### Найдите количество курящих людей

```python
counts = df['smoker'].value_counts()
smoker_yes = counts.get('Yes')
smoker_yes
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
mean_total_bill_by_day = df.groupby('day')['total_bill'].mean()
mean_total_bill_by_day
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
mediana = df['total_bill'].median()
filter_total_bill = df.query('total_bill > @mediana')

sex_mean = filter_total_bill.groupby(['sex'])['tip'].mean()
sex_mean
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
df['smoker'] = df['smoker'].replace({'No': 0, 'Yes': 1})
df
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
sns.histplot(data=df, x='total_bill')
plt.title('Распределение признака "total_bill"')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.scatter(df['total_bill'],df['tip'])
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('Диаграмма рассеивания')
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(df, hue='size')

plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
sns.relplot(data=df, x='day', y='total_bill')
plt.xlabel('День недели')
plt.ylabel('Суммарный счет')
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
fig, ax = plt.subplots(figsize=(12, 6))

for time in ['Lunch', 'Dinner']:
    subset = df.query(f"time == '{time}'")['tip']
    sns.histplot(subset, label=time, kde=False, ax=ax)

plt.legend(title='Время приема пищи')
plt.title('Гистограмма распределения чаевых в зависимости от времени приема пищи')
plt.xlabel('Размер чаевых (USD)')
plt.ylabel('Количество наблюдений')
plt.tight_layout()
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
# Создаём фигуру и оси для двух графиков
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Построение графика для мужчин
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=df[df["sex"] == "Male"], ax=axes[0])
axes[0].set_title("Взаимосвязь между total_bill и tip для мужчин")
axes[0].set_xlabel("Total bill")
axes[0].set_ylabel("Tip")

# Построение графика для женщин
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=df[df["sex"] == "Female"], ax=axes[1])
axes[1].set_title("Взаимосвязь между total_bill и tip для женщин")
axes[1].set_xlabel("Total bill")
axes[1].set_ylabel("Tip")

# Отображаем легенду
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=2)

plt.tight_layout()  # Убираем лишнее пространство вокруг графиков
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Путь к файлу на Github
path = 'https://raw.githubusercontent.com/Deniska400/DZ_OTUS/main/DZ_9/dataset.csv'
# Загрузка данных
df = pd.read_csv(path)
df = df.drop(columns=['id'])
df.head()
```

```python
df = df.dropna(axis=1, how='all')
df['diagnosis'] = df['diagnosis'].replace({'M': 1, 'B': 0})
df['diagnosis'].unique()
df['diagnosis'].value_counts(normalize=True)
```

```python
# описательная статистика
df.describe()
```

```python

for column in df.columns[1::]:
    a = df[column]
    ax = sns.histplot(data=df, x=column, hue='diagnosis', kde=True, palette='Set1', multiple='stack')

    medians = df.groupby('diagnosis')[column].median()
    for target_value, median_value in medians.items():
      ax.axvline(median_value, color='red' if target_value == 0 else 'blue', linestyle='--', label=f'Медиана (target ={target_value}) = {median_value:.2f}')

    plt.title(f'Распределение признака по "{column}" по целевой переменной')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.legend()
    plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(12,12))
sns.heatmap(corr, annot=False, cmap='coolwarm', vmin=-1, vmax=1,)
plt.title('Матрица корреляций')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x = df.drop(columns=['diagnosis'])
y = df['diagnosis']
x.shape, y.shape  # Размер X (количество строк, количество признаков),  Размер y (количество строк)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaler = scaler.fit_transform(x_test)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, roc_auc_score
```

```python
neighbors = range(1, 50)
f1_score_train = []
f1_score_test = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train_scaler, y_train)
    f1_score_train.append(f1_score(knn.predict(x_train_scaler), y_train))
    f1_score_test.append(f1_score(knn.predict(x_test_scaler), y_test))

plt.plot(neighbors, f1_score_train, color='blue', label='train')
plt.plot(neighbors, f1_score_test, color='red', label='test')
plt.title("Max test quality: {:.3f}\nBest k: {}".format(max(f1_score_test), np.argmax(f1_score_test)+1))
plt.legend();
```

```python
# Предсказания на тестовой выборке
y_pred = knn.predict(x_test)
y_pred_proba = knn.predict_proba(x_test)[:, 1]  # Вероятности для ROC-AUC

# Оценка качества модели
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)
print("Метрики качества модели kNN 'из коробки':")
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")
print(f"ROC-AUC: {roc_auc:.3f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Параметры для настройки
param_grid = {'n_neighbors': range(1, 50)}  # Перебираем k от 1 до 49

# Создание модели с кросс-валидацией
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='f1')
grid_search.fit(x_train, y_train)

# Лучший параметр
best_k = grid_search.best_params_['n_neighbors']
print(f"Оптимальное число соседей: {best_k}")

# Обучение модели с лучшим параметром
knn_tuned = KNeighborsClassifier(n_neighbors=best_k)
knn_tuned.fit(x_train, y_train)

# Предсказания на тестовой выборке
y_pred_tuned = knn_tuned.predict(x_test)
y_pred_proba_tuned = knn_tuned.predict_proba(x_test)[:, 1]  # Вероятности для ROC-AUC

# Оценка качества модели
accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
precision_tuned = precision_score(y_test, y_pred_tuned)
recall_tuned = recall_score(y_test, y_pred_tuned)
f1_tuned = f1_score(y_test, y_pred_tuned)
roc_auc_tuned = roc_auc_score(y_test, y_pred_proba_tuned)

print("\nМетрики качества модели kNN с настройкой параметра:")
print(f"Accuracy: {accuracy_tuned:.3f}")
print(f"Precision: {precision_tuned:.3f}")
print(f"Recall: {recall_tuned:.3f}")
print(f"F1-score: {f1_tuned:.3f}")
print(f"ROC-AUC: {roc_auc_tuned:.3f}")

# Построение ROC-кривой
fpr_tuned, tpr_tuned, thresholds_tuned = roc_curve(y_test, y_pred_proba_tuned)
plt.plot(fpr_tuned, tpr_tuned, label=f'ROC curve (AUC = {roc_auc_tuned:.3f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая (с настройкой параметра)')
plt.legend()
plt.show()
```
